```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
path_to_csv = "../data/train_data.csv"
data = pd.read_csv(path_to_csv, index_col=0)
```

```python
train_data, val_data = train_test_split(
    data,
    test_size=0.33,
    random_state=42,
)
```

```python
train_data.to_csv("../data/split_data/train_data.csv")
val_data.to_csv("../data/split_data/val_data.csv")
```
